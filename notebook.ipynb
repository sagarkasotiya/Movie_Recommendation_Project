```python
import ast
def genre_convert(obj):
    m=ast.literal_eval(obj)
    list=[]
    for i in m:
        list.append(i['name'])
    return list
```

```python
movies['genres']=movies['genres'].apply(genre_convert)
```

```python
def cast_convert(obj):
    m=ast.literal_eval(obj)
    list=[]
    count=0
    for i in m:
        if count!=5:
            list.append(i['name'])
            count=count+1
            
        else:
            break
    return list
```

```python
movies['cast']=movies['cast'].apply(cast_convert)
```

```python
import numpy as np
def fetch_director(obj):
    
    m=ast.literal_eval(obj)
    list=[]
    for i in m:
        if i['job']=='Director':
            list.append(i['name'])
    return list
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python

def space_remove(obj):
    for i in range(len(obj)):
        obj[i] = obj[i].replace(' ', '')
    return obj


```

```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob

def text_preprocessing(txt):
    #text to lower case
    txt=txt.lower()
    
    #removing punctuation
    translator=str.maketrans("","",string.punctuation)
    txt=txt.translate(translator)
    
    #remove digits
    txt="".join([word for word in txt if not word.isdigit()])
    
    #text without whitespace
    # txt=" ".join(txt.split())
    
    #txt normalization
    # txt=str(TextBlob(txt).correct())
    
    
    #word tokenize
    txt=word_tokenize(txt) #list

    #stop words removal
    stop_words=set(stopwords.words('english'))
    filtered_txt=[word for word in txt if word not in stop_words]
    
    #stemming
    stemmer=PorterStemmer()
    txt=[stemmer.stem(word) for word in filtered_txt]
    
    #lemmatization
    lemmatizer=WordNetLemmatizer()
    txt=[lemmatizer.lemmatize(word) for word in txt]
    
    return txt
```

```python
import swifter
movies['tags']=movies['tags'].swifter.apply(text_preprocessing)
movies.head()
```

```python
movies['tags']=movies['tags'].apply(lambda x: " ".join(word for word in x))
movies.head()
```

```python
#vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer=TfidfVectorizer(max_features=5000)
vectors=vectorizer.fit_transform(movies['tags'])
vectorizer.get_feature_names_out()
```

```python
#cosine distance
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
l=sorted(list(enumerate([float(dist) for dist in list(similarity[0])])),reverse=True,key=lambda x:x[1])
l
```

```python
def recommend(movie):
    movie_index=int(movies[movies['title']==movie].index[0])
    l=sorted(list(enumerate([float(dist) for dist in list(similarity[movie_index])])),reverse=True,key=lambda x:x[1])
    movie_indices=[]
    for touple in l[0:5]:
        movie_indices.append(touple[0])
    movies_id=[]
    for index in movie_indices:
        movies_id.append(movies['movie_id'][index])
    
    id=[int(id) for id in movies_id]
    return id
    
```
